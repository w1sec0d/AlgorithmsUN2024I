```python
# fibR - Fibonacci Recursive O( ( (1+sqrt(5)/2 )^n lg n ) [Relaxed asymptotic bound O(2^n lg n)] 
def fibR(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)    # for large nÂ´s  the addition will take lg n stetp because the numebers are very large
```

```python
#Plot of the recursive Fibonacci execution time for n in [0,45]
import time
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(46)
y = []
for n in x:
    start_time = time.time()
    fibR(n)
    y.append(time.time()-start_time)
plt.plot(x, y)
#Adds y=n to the plot
plt.plot(x, x)
plt.legend(['Recursive Fibonacci','y=n'])
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Recursive Fibonacci')
plt.show()

```

```python
# fibI - Fibonacci Iterartive (forward) O(n lgn)
def fibI(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```python
#Plot of the iterative Fibonacci execution time for n in [0,1000]
import time
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1001)
y = []
for n in x:
    start_time = time.time()
    fibI(n)
    y.append(time.time()-start_time)
plt.plot(x, y)
#Adds y=n to the plot
plt.plot(x, x)
plt.legend(['Iterative Fibonacci','y=n'])
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Iterative Fibonacci')
plt.show()

```

```python
# fibI - Fibonacci Memoization (dynamic programming) (backward) O(n lg n)

def fibM(n):
    memo = {}
    def fib(n):
        if n==0 :
            return 0
        if n==1 :
            return 1
        if (n-2) not in memo:
            memo[n-2] = fib(n-2)
        if (n-1) not in memo:
            memo[n-1] = fib(n-1)
        return memo[n-1] + memo[n-2]
    return fib(n)
```

```python
#Plot of the logarithmic Fibonacci execution time for n in [0,1000]
import time
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1001)
y = []
for n in x:
    start_time = time.time()
    fibM(n)
    y.append(time.time()-start_time)
plt.plot(x, y)
#Adds y=n to the plot
plt.plot(x, x)
plt.legend(['Logarithmic Fibonacci','y=n'])
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Logarithmic Fibonacci')
plt.show()

```
